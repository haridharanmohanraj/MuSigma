```python
# Step 4: Fill the null values in the specified columns with 'unknown'
df['ReturnReason'] = df['ReturnReason'].fillna('NOT pROVIDED')
df['CrossSellProducts'] = df['CrossSellProducts'].fillna('no')
df['UpsellProducts'] = df['UpsellProducts'].fillna('unknown')
```

```python
# Step 5: Check for duplicates in the main data
duplicates = df.duplicated(subset=df.columns).sum()
print(f"Total Duplicates: {duplicates}")
```

```python
# 2. Check for Duplicates (No deletion, add surrogate key if needed)
df["SurrogateID"] = range(1, len(df) + 1)  # Unique identifier
```

```python
# 3. Create Dimension Tables with Surrogate Keys

# Customer Dimension Table
dim_customer = df[[
    "CustomerID", "CustomerAge", "CustomerGender", "CustomerLocation", 
    "LoyaltyProgramMember", "CustomerFeedbackRating"
]].drop_duplicates().reset_index(drop=True)
dim_customer["CustomerKey"] = dim_customer.index + 1  # Surrogate Key

# Product Dimension Table
dim_product = df[[
    "ProductID", "ProductName", "ProductCategory"
]].drop_duplicates().reset_index(drop=True)
dim_product["ProductKey"] = dim_product.index + 1  # Surrogate Key

# Store Dimension Table
dim_store = df[[
    "StoreLocation", "Region"
]].drop_duplicates().reset_index(drop=True)
dim_store["StoreKey"] = dim_store.index + 1  # Surrogate Key

# Shipping Dimension Table
dim_shipping = df[[
    "ShippingMethod", "ShippingFee", "ShippingStatus"
]].drop_duplicates().reset_index(drop=True)
dim_shipping["ShippingKey"] = dim_shipping.index + 1  # Surrogate Key

# Promotion Dimension Table
dim_promotion = df[[
    "PromoCodeUsed", "DiscountPercentage"
]].drop_duplicates().reset_index(drop=True)
dim_promotion["PromotionKey"] = dim_promotion.index + 1  # Surrogate Key

# Sales Strategy Dimension Table (For Cross-Sell & Up-Sell)
dim_sales_strategy = df[[
    "CrossSellProducts", "UpsellProducts"
]].drop_duplicates().reset_index(drop=True)
dim_sales_strategy["SalesStrategyKey"] = dim_sales_strategy.index + 1  # Surrogate Key

# Transaction Details Dimension Table
dim_transaction_details = df[[
    "TimeOfDay", "SalesChannel", "PaymentMethod", "PaymentStatus", "TransactionStatus", "ReturnFlag", "ReturnReason"
]].drop_duplicates().reset_index(drop=True)
dim_transaction_details["TransactionDetailKey"] = dim_transaction_details.index + 1  # Surrogate Key

# 4. Merge Surrogate Keys into Fact Table
fact_sales = df.merge(dim_customer, on=["CustomerID", "CustomerAge", "CustomerGender", "CustomerLocation", 
                                        "LoyaltyProgramMember", "CustomerFeedbackRating"], how="left")

fact_sales = fact_sales.merge(dim_product, on=["ProductID", "ProductName", "ProductCategory"], how="left")
fact_sales = fact_sales.merge(dim_store, on=["StoreLocation", "Region"], how="left")
fact_sales = fact_sales.merge(dim_shipping, on=["ShippingMethod", "ShippingFee", "ShippingStatus"], how="left")
fact_sales = fact_sales.merge(dim_promotion, on=["PromoCodeUsed", "DiscountPercentage"], how="left")
fact_sales = fact_sales.merge(dim_sales_strategy, on=["CrossSellProducts", "UpsellProducts"], how="left")
fact_sales = fact_sales.merge(dim_transaction_details, on=["TimeOfDay", "SalesChannel", "PaymentMethod", "PaymentStatus", 
                                                           "TransactionStatus", "ReturnFlag", "ReturnReason"], how="left")

# Select only relevant columns for fact table (only numerical attributes)
fact_sales = fact_sales[[
    "SurrogateID", "TransactionID", "CustomerKey", "ProductKey", "StoreKey", "ShippingKey", "PromotionKey", "SalesStrategyKey", "TransactionDetailKey",
    "SalesQuantity", "SalesPrice", "DiscountAmount", "FinalSalePrice", "TransactionDate", "SalesTax"
]]
```
