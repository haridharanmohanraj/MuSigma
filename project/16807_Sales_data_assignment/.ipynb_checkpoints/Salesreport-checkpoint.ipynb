#  Importing the dataset


```python
import pandas as pd
data = pd.read_csv("Sales Data.csv")  
data.head()
```

# Check for Null Values

```python
data.isnull().sum()
```

```python
#Fill the null values in the specified columns 
data['ReturnReason'] = data['ReturnReason'].fillna('Not Provided')
data['CrossSellProducts'] = data['CrossSellProducts'].fillna('No Product')
data['UpsellProducts'] = data['UpsellProducts'].fillna('No Product')
```

```python
data.isnull().sum()
```

```python
# data = data.drop('ReturnFlag', axis=1)
data = data.drop_duplicates(subset='TransactionID', keep='last')
```

# Create Dimensional Tables

```python
customer = data[[
"CustomerID", "CustomerAge", "CustomerGender", "CustomerLocation", "LoyaltyProgramMember", "CustomerFeedbackRating"
]].drop_duplicates().reset_index(drop=True)

customer["CustomerIndex"] = "C" + (customer.index + 1).astype(str)
customer
```

```python
product = data[[
    "ProductID", "ProductName", "ProductCategory","CrossSellProducts","UpsellProducts"
]].drop_duplicates().reset_index(drop=True)

product["ProductIndex"] = "P"  + (product.index + 1).astype(str)
product
```

```python
store = data[[
    "StoreLocation", "Region"
]].drop_duplicates().reset_index(drop=True)

store["StoreIndex"] = "ST" + (store.index + 1).astype(str)
store
```

```python

Transaction = data[[
   'TransactionDate', 'TimeOfDay','SalesChannel'
]].drop_duplicates().reset_index(drop=True)

Transaction["TransactionIndex"] = "T" + (Transaction.index + 1).astype(str)
Transaction
```

```python
payment = data[[
   'PaymentMethod','PaymentStatus','TransactionStatus'
]].drop_duplicates().reset_index(drop=True)

payment["paymentIndex"] = "P" + (payment.index + 1).astype(str)
payment
```

```python
ship = data[[
   'ShippingMethod','ShippingStatus'
]].drop_duplicates().reset_index(drop=True)

ship["ShipIndex"] = "SH" + (ship.index + 1).astype(str)
ship
```

```python
## Review the dim table 
```

```python
ship.info()
payment.info()
Transaction.info()
customer.info()
product.info()
store.info()
```

# Join the dateset and dim table 

```python
Sales = data.copy()


Sales = Sales.merge(customer, on=["CustomerID", "CustomerAge", "CustomerGender", "CustomerLocation", "LoyaltyProgramMember", "CustomerFeedbackRating"], how="left")
Sales = Sales.merge(product, on=["ProductID", "ProductName", "ProductCategory","CrossSellProducts","UpsellProducts"], how="left")
Sales = Sales.merge(store, on=["StoreLocation", "Region"], how="left")
Sales = Sales.merge(ship, on=["ShippingMethod", "ShippingStatus"], how="left")
Sales = Sales.merge(payment, on=['PaymentMethod','PaymentStatus','TransactionStatus'], how="left")
Sales = Sales.merge(Transaction, on=['TransactionDate', 'TimeOfDay','SalesChannel'], how="left")
```

```python
Sales_Fact = Sales[["TransactionIndex","ProductIndex","StoreIndex","paymentIndex","ShipIndex","CustomerIndex","TransactionID","SalesTax","SalesQuantity", "SalesPrice", "DiscountAmount", "FinalSalePrice","PromoCodeUsed", "DiscountPercentage", "ShippingFee"]]
```

## Export the file in excel format

```python
customer.to_excel("Customer.xlsx",index=False)
product.to_excel("product.xlsx",index=False)
ship.to_excel("ship.xlsx",index=False)
Transaction.to_excel("Transaction.xlsx",index=False)
store.to_excel("Store.xlsx",index=False)
payment.to_excel("payment.xlsx",index=False)
Sales_Fact.to_excel("Sales.xlsx",index=False)
```
